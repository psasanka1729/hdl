```julia
Nx=1 
Ny_max=2

U11=1 # nearest neighbor potential in first layer
U22=1 # nearest neighbor potential term in 2nd layer
Usi=1# nearest neighbor potential betweeen hydrogen and silicon atom

t22=1 # nearest nighbor hopping in 2nd layer
t11=1 # nearest neighbor hopping in first layer
t_si= # nearest neighor hopping between silicon and hydrogen atom
t_b1=1 # loss/gain at boundary 1
t_b2=2 # loss/gain at boundary 2
t_b3=3 # loss/gain at boundary 3
t_b4=4; # loss/gain at boundary 4
```

## All possible states

```julia
it=Iterators.product(ntuple(_ -> 0:1, 1+Nx*Ny_max*6)...);
p=collect(it) # possible states with non conserving particle number
```

# The Hamiltonian

```julia
# Generating a matrix with zeros.
H=zeros(2^(1+Nx*Ny_max*6),2^(1+Nx*Ny_max*6)); # 6 nearest neighbours.
size(H)
```

## 1 Nearest Neighbor potential first main layer potential

```julia
for n=1:2^(1+Nx*Ny_max*6)
    U=0;
    for Ny=1:Ny_max

        for k=6*(Nx)*(Ny-1)+1:6*(Nx)*(Ny-1)+4*Nx-1
            println(k);
            println(k+1);
            if p[n][k]==1 && p[n][k+1]==1
                U=U+U11;
            end
        end
    end
    H[n,n]+=U;
end
```

## 2 Nearest neighbor potential term between hydrogen and silicon atom at the center

```julia
U_si=0
for n=1:2^(1+Nx*Ny_max*6)
    if Nx>1
        nx=Int(Nx/2)
    else nx=1
    end
    
    if Ny_max>2
        ny=Int(Ny_max/2)
    else        ny=1
    end
    
    if p[n][1+Nx*Ny_max*6]==1 && p[n][ny*nx*6]==1
        U_si=Usi
    end
    H[n,n]+=U_si
    
end
```

## 3 Nearest Neighbor potential 2nd main layer potential

```julia
for n=1:2^(1+Nx*Ny_max*6)
    U=0
    for Ny=1:Ny_max
        for k=6*(Nx)*(Ny-1)+4*Nx+1:2:6*(Nx)*(Ny-1)+4*Nx+2*Nx-1
            println(k)
            println(k+1)
            if p[n][k]==1 && p[n][k+1]==1
                U=U+U22
            end
        end
    end
    H[n,n]+=U
end
```

## 4 Nearest neighbor potential from first to 2nd layer

```julia
for n=1:2^(1+Nx*Ny_max*6)
    U=0
    for Ny=1:Ny_max
        for n=2:2:4*Nx
            #println(Nx*6*(Ny-1)+n)
            #println(Nx*6*(Ny-1)+4*Nx+Int(n/2))
            if p[n][Nx*6*(Ny-1)+n]==1 && p[n][Nx*6*(Ny-1)+4*Nx+Int(n/2)]==1
                U=U+U12
            end
        end
    end
    H[n,n]+=U
end
```

## 5 Nearest neighbor potential from 2nd layer to 3rd layer

```julia
for n=1:2^(1+Nx*Ny_max*6)
    U=0
    for Ny=2:Ny_max
        for n1=2:2:4*Nx
            #println(Nx*6*(Ny-1)+n1)
            #println(Nx*6*(Ny-2)+4*Nx+Int(n1/2))
            if p[n][Nx*6*(Ny-1)+n1]==1 && p[n][Nx*6*(Ny-1)+4*Nx+Int(n1/2)]==1
                U=U+U23
            end
        end      
    end
    H[n,n]+=U    
end
```

# Hopping

## 6 Nearest neighbor hopping between hydrogen and silicon atom

```julia
for n=1:2^(1+Nx*Ny_max*6)
    
    if Nx>1
        
        nx=Int(Nx/2)
        
    else nx=1
        
    end
    
    if Ny_max>2
        
        ny=Int(Ny_max/2)
        
    else        ny=1
        
    end
            if p[n][1+Nx*Ny_max*6]==1 && p[n][ny*nx*6]==0

                q=p[n]
                q=collect(p[n])
                q[1+Nx*Ny_max*6]=0
                q[ny*nx*6]=1
                phase1=(-1)^(sum(p[n][1+Nx*Ny_max*6:1+Nx*Ny_max*6]))
                phase2=1
        
                if 1+Nx*Ny_max*6<Nx*Ny_max*6
                    phase2=(-1)^(sum(p[n][ny*nx*6:6*Nx*Ny_max+1]))
                end
        
                for n1=1:2^(Nx*Ny_max*6)
                    if collect(p[n1])==q
                        #                         push!(U0,n1)
                        H[n,n1]=H[n,n1]+t_si*phase1*phase2
                    end
                end
            end
end
```

## 7 Nearest Neighbor hopping first main layer 

```julia
for n=1:2^(1+Nx*Ny_max*6)
    for Ny=1:Ny_max
        
        for k=6*(Nx)*(Ny-1)+1:6*(Nx)*(Ny-1)+4*Nx-1
            #println(k)
            #println(k+1)
            if p[n][k]==1 && p[n][k+1]==0
                q=p[n]
                q=collect(p[n])
                q[k]=0
                q[k+1]=1
                phase1=(-1)^(sum(p[n][k:1+Nx*Ny_max*6]))
                phase2=1
                if k+1<Nx*Ny_max*6
                    phase2=(-1)^(sum(p[n][1+k:6*Nx*Ny_max+1]))
                end
                for n1=1:2^(Nx*Ny_max*6)
                    if collect(p[n1])==q
                        #                         push!(U0,n1)
                        H[n,n1]=H[n,n1]+t11*phase1*phase2
                    end
                end
            end
            
        end
    end
end
```

## 8 Nearest Neighbor hopping 2nd main layer

```julia
for n=1:2^(1+Nx*Ny_max*6)
    
    for Ny=1:Ny_max
        
        for k=6*(Nx)*(Ny-1)+4*Nx+1:2:6*(Nx)*(Ny-1)+4*Nx+2*Nx-1
            #println(k)
            #println(k+1)
            if p[n][k]==1 && p[n][1+k]==0   
                q=p[n]
                q=collect(p[n])
                q[k]=0
                q[1+k]=1
                phase1=(-1)^(sum(p[n][k:6*Nx*Ny_max+1]))
                phase2=1
                if 1+k<6*Nx*Ny_max+1
                    phase2=(-1)^(sum(p[n][1+k:6*Nx*Ny_max+1]))
                end
                for n1=1:2^(6*Nx*Ny_max+1)
                    if collect(p[n1])==q
                        H[n,n1]=H[n,n1]+t22*phase1*phase2
                    end
                end
                
            end
        end
    end
end
```

```julia
for n=1:2^(1+Nx*Ny_max*6)
    
    for Ny=1:Ny_max
        for n=2:2:4*Nx
            #println(Nx*6*(Ny-1)+n)
            #println(Nx*6*(Ny-1)+4*Nx+Int(n/2))
            
            if p[n][Nx*6*(Ny-1)+n]==1 && p[n][Nx*6*(Ny-1)+4*Nx+Int(n/2)]==0   
                q=p[n]
                q=collect(p[n])
                q[Nx*6*(Ny-1)+n]=0
                q[Nx*6*(Ny-1)+4*Nx+Int(n/2)]=1
                phase1=(-1)^(sum(p[n][Nx*6*(Ny-1)+n:6*Nx*Ny_max+1]))
                phase2=1
                if Nx*6*(Ny-1)+4*Nx+Int(n/2)<6*Nx*Ny_max+1
                    phase2=(-1)^(sum(p[n][Nx*6*(Ny-1)+4*Nx+Int(n/2):6*Nx*Ny_max+1]))
                end
                for n1=1:2^(6*Nx*Ny_max+1)
                    if collect(p[n1])==q
                        H[n,n1]=H[n,n1]+t12*phase1*phase2
                    end
                end
                
            end
            
        end
        
    end
end
```

## 10 Nearest neighbor potential from 2nd layer to 3rd layer

```julia
for n=1:2^(1+Nx*Ny_max*6)
    
    for Ny=2:Ny_max
        for n=2:2:4*Nx
            #println(Nx*6*(Ny-1)+n)
            #println(Nx*6*(Ny-2)+4*Nx+Int(n/2))
            
            if p[n][Nx*6*(Ny-1)+n]==1 && p[n][Nx*6*(Ny-2)+4*Nx+Int(n/2)]==0   
                q=p[n]
                q=collect(p[n])
                q[Nx*6*(Ny-1)+n]=0
                q[Nx*6*(Ny-2)+4*Nx+Int(n/2)]=1
                phase1=(-1)^(sum(p[n][Nx*6*(Ny-1)+n:6*Nx*Ny_max+1]))
                phase2=1
                if Nx*6*(Ny-2)+4*Nx+Int(n/2)<6*Nx*Ny_max+1
                    phase2=(-1)^(sum(p[n][Nx*6*(Ny-2)+4*Nx+Int(n/2):6*Nx*Ny_max+1]))
                end
                for n1=1:2^(6*Nx*Ny_max+1)
                    if collect(p[n1])==q
                        H[n,n1]=H[n,n1]+t23*phase1*phase2
                    end
                end
                
            end
        end
        
    end
end
```

```julia
for n=1:2^(Nx*Ny_max*6+1)
    for Ny=1:1
        
        for k=6*(Nx)*(Ny-1)+1:6*(Nx)*(Ny-1)+4*Nx-1
            #println(k)
            #println(k+1)
            if p[n][k]==1 
                q=p[n]
                q=collect(p[n])
                q[k]=0
                # phase1=(-1)^(sum(p[n][k:Nx*Ny_max*6]))
                phase2=1
                if k<Nx*Ny_max*6
                    phase2=(-1)^(sum(p[n][k:6*Nx*Ny_max+1]))
                end
                for n1=1:2^(1+Nx*Ny_max*6)
                    if collect(p[n1])==q
                        #                         push!(U0,n1)
                        H[n,n1]=H[n,n1]+t_b1*phase2
                    end
                end
            end
            
        end
    end
end
```
